```python
import pandas as pd
```

```python
df = pd.read_csv ('../vine_table.csv')
df.head(10)
```

## 1. Create new dataframe where total_votes is equal or greater than 20

```python
total_votes_df = df.loc[df['total_votes'] >= 20]
total_votes_df.head(10)
```

## 2. Create new dataframe from step 1 where helpful_votes divided by total_votes is equal to or greater than 50%

```python
total_votes_df['helpful_votes_percent'] = total_votes_df['helpful_votes'] / total_votes_df['total_votes']
high_helpful_votes_df = total_votes_df.loc[total_votes_df['helpful_votes_percent'] >= .5]
high_helpful_votes_df.head(10)
```

## 3. Create new dataframe from step 2 where a review was written as part of the Vine program (paid), vine == 'Y' 

```python
paid_review_df = high_helpful_votes_df.loc[high_helpful_votes_df['vine'] == 'Y']
paid_review_df.head(5)
```

## 4. Create new dataframe from step 2 where the review was not part of the Vine program (unpaid), vine == 'N'

```python
unpaid_review_df = high_helpful_votes_df.loc[high_helpful_votes_df['vine'] == 'N']
unpaid_review_df.head(5)
```

## 5. Determine the total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews for the two types of review (paid vs unpaid).

```python
#paid: Total number of reviews
len(paid_review_df)
```

```python
#paid: number of 5 star reviews
len(paid_review_df[paid_review_df['star_rating'] == 5])
```

```python
#paid: percentage of 5 star reviews
percentage = len(paid_review_df[paid_review_df['star_rating'] == 5]) / len(paid_review_df)
print(f'{percentage*100:.2f}%')
```

```python
#unpaid: total number of reviews
len(unpaid_review_df)
```

```python
#unpaid: number of 5 star reviews
len(unpaid_review_df[unpaid_review_df['star_rating'] == 5])
```

```python
#unpaid: percentage of 5 star reviews
percentage_unpaid = len(unpaid_review_df[unpaid_review_df['star_rating'] == 5]) / len(unpaid_review_df)
print(f'{percentage_unpaid*100:.2f}%')
```
